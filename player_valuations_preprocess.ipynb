```python
import pandas as pd
```

```python
comp = pd.read_csv('competitions.csv')
comp.head()
```

```python
pv = pd.read_csv('player_valuations.csv').drop(['dateweek','datetime','current_club_id',
                                            'n'], axis=1)
pv.head()
```

```python
p = pd.read_csv('players.csv').drop(['market_value_in_eur','last_season','current_club_id','first_name',
                                  'last_name','city_of_birth','url',
                                     'country_of_citizenship','agent_name','image_url',
                                     'highest_market_value_in_eur'], axis=1)
new = pd.merge(p, pv, on=['player_id'])
new['date'] = pd.to_datetime(new['date'])
new = new.sort_values(['player_id', 'date']).drop_duplicates('player_id', keep='last')
new['Position'] = new['position'].replace({'Attack': 'Forward', 'Midfield': 'Midfielder'})
new = new.drop(['position'],axis=1)
new[new['player_code']=='rasmus-kristensen']
```

```python
new.to_csv('transfermarkt_data.csv')
```
